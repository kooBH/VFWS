ID로 폴더를 만들어서 그안에다가 정규화한 파일들을 넣어 두자.

```python
import multiprocessing as mp
from pydub import AudioSegment, effects  
from tqdm import tqdm

def get_file_name(x) :
    x = x.split('/')
    return x[-1]
```

```python
norm_dir = '/home/nas/user/kbh/VFWS/norm/'
num_core = mp.cpu_count()
print('num core : ' + str(num_core))

if not os.path.isdir(norm_dir) :
    os.mkdir(norm_dir)

# normalize target wav [mldr], clean
def norm_target(arg):
    path = norm_dir + str(df.iloc[arg].name) +'/'
    if not os.path.isdir(path) :
        os.mkdir(path)
    for x in df.iloc[arg] :
        if isinstance(x, list):
            for xx in x :
                rawsound = AudioSegment.from_wav(xx)
                xx_name = get_file_name(xx)
                normalizedsound = effects.normalize(rawsound)  
                normalizedsound.export(path + xx_name, format="wav")
                print('normalized : ' + xx_name)
        else : 
            rawsound = AudioSegment.from_wav(x)
            x_name = get_file_name(x)
            normalizedsound = effects.normalize(rawsound)  
            normalizedsound.export(path + x_name, format="wav")
            print('normalized : ' + x_name)
    return     

pool = mp.Pool(processes=num_core)
pool.map(norm_target, range(len(df)))

pool.close()
pool.join()
```

# (2)싱크

norm  폴더에 대해서 각각의 ID를 폴더명으로 가지는 하위 디렉터리에 대해 for문을 돌려서 각 폴더에내서 싱크를 맞추자.

```python
import wav_syncer
import os
import glob
from tqdm import tqdm
import multiprocessing as mp

num_core = mp.cpu_count()
print('num core : ' + str(num_core))

norm_dir = '/home/nas/user/kbh/VFWS/norm/'
list_norm = os.listdir(norm_dir)
sync_dir = '/home/nas/user/kbh/VFWS/sync/'
targets = ['_BUS.wav','_CAF.wav','_PED.wav','_STR.wav']

def sync_norm_wav(idx) : 
    x = list_norm[idx]
    print('sycnronizaing : ' + x)
    target_path = norm_dir + x
    sync_sub_dir = sync_dir + x
    if not os.path.isdir(sync_sub_dir) :
        os.mkdir(sync_sub_dir)
    
    #print(target_path)
    if os.path.isdir(target_path) : 
        #rint(target_path)
        base = target_path + '/'+x
        anchor = base+'_BTH.CH0.wav'
        

        for t in targets : 
            tt = base + t
            wav_syncer.sync_wav(anchor,tt,sync_sub_dir+'/'+x+'_BTH'+t,sync_sub_dir+'/'+x+t)
            
pool = mp.Pool(processes=num_core)
pool.map(sync_norm_wav, range(len(list_norm)))

pool.close()
pool.join()       
        
```
